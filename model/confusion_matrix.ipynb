```python
import numpy as np
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = np.array([(60,2,3,6), (12,81,6,7), (10,5,78,8), (8,24,17,50)])
```

```python
display_labels = ['Angry', 'Happy', 'Sad', 'Relaxed']
cm_display = ConfusionMatrixDisplay(cm, display_labels=display_labels).plot(cmap = 'Blues')
```
